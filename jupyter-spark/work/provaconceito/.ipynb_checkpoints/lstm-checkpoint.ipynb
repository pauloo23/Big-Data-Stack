## Helper Functions
Funções para converter as séries temporais para supervised de modo a serem alimentadas no modelo LSTM.

```python
# multi-step data preparation
from numpy import array

# split a univariate sequence into samples
def split_sequence(sequence, n_steps_in, n_steps_out):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps_in
        out_end_ix = end_ix + n_steps_out
        # check if we are beyond the sequence
        if out_end_ix > len(sequence):
            break
        # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix:out_end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

```python
from rpy2.robjects.packages import importr
import rpy2.robjects.numpy2ri
from rpy2.robjects import pandas2ri,r
from rpy2.robjects.conversion import localconverter
rpy2.robjects.numpy2ri.activate()
pandas2ri.activate()
metrics = r("""r2 = function(ytrue, y_pred){
  obj = rminer::mmetric(unlist(ytrue), unlist(y_pred), metric=c("R2", "MAE", "MSE", "RMSE"))
  return (obj)
}""")
```

# Encoder-Decoder Model

```python
ytrain2 = ytrain.copy()
ytrain2 = ytrain2.reshape((ytrain2.shape[0], ytrain2.shape[1], n_features))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import RepeatVector
from tensorflow.keras.layers import TimeDistributed
 
# define model
model2 = Sequential()
model2.add(LSTM(100, activation=tf.keras.layers.LeakyReLU(), input_shape=(n_steps_in, n_features)))
model2.add(RepeatVector(n_steps_out))
model2.add(LSTM(100, activation=tf.keras.layers.LeakyReLU(), return_sequences=True))
model2.add(TimeDistributed(Dense(1)))
model2.compile(optimizer='adam', loss='mse')
```

```python
# fit model
model2.fit(Xtrain, ytrain2, epochs=50, verbose=2)
```

# Normal LSTM

```python
# define model
model3 = Sequential()
model3.add(LSTM(100, activation=tf.keras.layers.LeakyReLU(), input_shape=(n_steps_in, n_features)))
model3.add(Dense(n_steps_out))
model3.compile(optimizer='adam', loss='mse')
```

```python
# fit model
model3.fit(Xtrain, ytrain, epochs=50, verbose=2)
```

```python
features = 1
# make one forecast with an LSTM normal!
def forecast_lstm(model, X, n_steps_in, n_features):
    # reshape input pattern to [samples, timesteps, features]
    X = X.reshape((1, n_steps_in, n_features))
    # make forecast
    forecast = model.predict(X)
    # convert to array
    return [x for x in forecast[0, :]]

def make_forecasts(model, testX, n_steps_in, features):
    forecasts = list()
    for i in trange(len(testX)):
        X,y = testX[i], ytest[i]
        
        # make forecast
        forecast = forecast_lstm(model, X, n_steps_in, features)
        # store the forecast
        forecasts.append(forecast)
    return forecasts
```

```python
forecasts = make_forecasts(model, Xtest, n_steps_in, 1)
```

```python
forecasts2 = make_forecasts(model2, Xtest, n_steps_in, 1)
```

```python
forecasts3 = make_forecasts(model3, Xtest, n_steps_in, 1)
```

# Vertical Evaluation

```python
# evaluate the RMSE for each forecast time step
def evaluate_forecasts(test, forecasts, n_seq, mode = "normal"):
    verticalEvaluation = pd.DataFrame(columns=["T","R2", "MAE", "MSE", "RMSE", "DatasetActual", "DatasetPred"])
    vertical = []
    for i in range(n_seq):
        
        actual = [row[i] for row in test]
        #print("Sample of ACTUAL SEQUENCE:", actual, "with total size of:", len(actual))
        if mode == "normal":
            predicted = [forecast[i] for forecast in forecasts]
        else:
            predicted = [forecast[i] for forecast in forecasts]
        #print("Sample of PREDICTED SIGNAL SEQUENCE:", predicted,"with total size of:", len(predicted))
        met = metrics(actual, predicted)
        r2 = met[0]
        MAE = met[1]
        MSE = met[2]
        RMSE = met[3]
        vertical2 = [[i+1, r2, MAE,MSE,RMSE, actual, predicted]]
        vertical = pd.DataFrame(vertical2, columns=["T","R2", "MAE", "MSE", "RMSE", "DatasetActual", "DatasetPred"])
        #add to original dataset
        verticalEvaluation = verticalEvaluation.append(vertical)
        print('t+%d - R^2: %f' % ((i+1), r2), '- MAE: %f' % (MAE), '- MSE: %f' % (MSE), '- RMSE: %f' % (RMSE))
    return verticalEvaluation
```
