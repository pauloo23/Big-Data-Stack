```python
dataframe = dataframe.select("sensor_created_at","sensor_event_type","sensor_id_cycle", \
                             "sensor_ip", "sensor_unique_id", "sensor_value_humidity", "sensor_value_temperature_motor1", "sensor_value_temperature_motor2", \
                            "sensor_value_temperature_motor3", "sensor_value_vibrationhz_x", "sensor_value_vibrationhz_y", "sensor_value_vibrationhz_z", \
                              "value_noise_dba_motor1", "value_noise_dba_motor2", "value_noise_dba_motor3")
```

```python
#dataframe.toPandas()
dataframe = dataframe.withColumn("sensor_value_humidity", dataframe["sensor_value_humidity"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_value_temperature_motor1", dataframe["sensor_value_temperature_motor1"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_value_temperature_motor2", dataframe["sensor_value_temperature_motor2"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_value_temperature_motor3", dataframe["sensor_value_temperature_motor3"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_value_vibrationhz_x", dataframe["sensor_value_vibrationhz_x"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_value_vibrationhz_y", dataframe["sensor_value_vibrationhz_y"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_value_vibrationhz_z", dataframe["sensor_value_vibrationhz_z"].cast(IntegerType()))
dataframe = dataframe.withColumn("value_noise_dba_motor1", dataframe["value_noise_dba_motor1"].cast(IntegerType()))
dataframe = dataframe.withColumn("value_noise_dba_motor2", dataframe["value_noise_dba_motor2"].cast(IntegerType()))
dataframe = dataframe.withColumn("value_noise_dba_motor3", dataframe["value_noise_dba_motor3"].cast(IntegerType()))
dataframe = dataframe.withColumn("sensor_created_at",dataframe['sensor_created_at'].cast(TimestampType()))
dataframe.toPandas()
```

```python
dataframe = dataframe.withColumn("temp_dayOfMonth", dayofmonth("sensor_created_at"))
dataframe = dataframe.withColumn("temp_month", month("sensor_created_at"))
dataframe = dataframe.withColumn("temp_hour", substring("sensor_created_at", 11, 13))
dataframe = dataframe.withColumn("temp_date", substring("sensor_created_at", 0, 10))
dataframe.toPandas()
```

```python
dataframe = dataframe.withColumn("temp_hour2", substring("sensor_created_at", 0, 16).cast(TimestampType())) \
                     .orderBy("temp_hour2")
dataframe.toPandas()
```

```python
dataframe2 = dataframe
```

```python
groupByColName = "sensor_ip"
colName = "sensor_value_humidity"

df3 = dataframe2.groupBy("temp_hour2", groupByColName) \
     .agg( \
     min(colName).alias('temp_min_hour'), \
     max(colName).alias('temp_max_hour'), \
     round(avg(colName), 2).alias('temp_avg_hour'), \
     count(colName).alias("temp_count_hour")) \
     .orderBy("temp_hour2").toPandas()
```
